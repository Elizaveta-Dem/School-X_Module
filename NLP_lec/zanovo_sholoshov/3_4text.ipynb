```python
class TextSimilarityCalculator: 
    def __init__(self): 
        self.vectorizer = CountVectorizer() 
         
    def calculate_similarity(self, text4): 
        # Преобразование текстов в векторы с помощью CountVectorizer 
        vectors = self.vectorizer.fit_transform(text4) 
         
        # Cравнение сходства текстов 
        similarity_matrix = cosine_similarity(vectors, dense_output=False) 
         
        return similarity_matrix
```

```python
similar_calculator = TextSimilarityCalculator() 
texts = [text3, text4]  # Список текстов
similar_matrix = similar_calculator.calculate_similarity(texts)
```

```python
for i, j in itertools.combinations(range(len(texts)), 2): 
    similarity = similar_matrix[i, j] 
    print(f"Косинусное сходство между текстом {i+1} и текстом {j+1}: {similarity}") 
```

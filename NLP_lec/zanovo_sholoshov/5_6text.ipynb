```python
class TextPreprocessor: 
    def __init__(self): 
        self.nlp = spacy.load("ru_core_news_sm") 
 
    def preprocess_texts(self, text5): 
        doc = self.nlp(text5) 
        words_filter = [token.lemma_ for token in doc if not token.is_punct and not token.is_stop] 
        preprocessed_text = " ".join(words_filter) 
        return preprocessed_text
```

```python
preprocessor_t = TextPreprocessor() 
preprocessed_text = preprocessor_t.preprocess_texts(text5) 
print(preprocessed_text)
```

```python

class TextSimilarityCalculator: 
    def __init__(self): 
        self.vectorizer = CountVectorizer() 
         
    def calculate_similarity(self, text6): 
        # Преобразование текстов в векторы с помощью CountVectorizer 
        vectors = self.vectorizer.fit_transform(text6) 
         
        # Cравнение сходства текстов 
        similarity_matrix = cosine_similarity(vectors, dense_output=False) 
         
        return similarity_matrix
```

```python
similar_calculator = TextSimilarityCalculator() 
texts = [text5, text6]  # Список текстов
similar_matrix = similar_calculator.calculate_similarity(texts)
```

```python
for i, j in itertools.combinations(range(len(texts)), 2): 
    similarity = similar_matrix[i, j] 
    print(f"Косинусное сходство между текстом {i+1} и текстом {j+1}: {similarity}") 
```

```python
class MyTokenizer:
    def __init__(self, stopwords=None):
        self.stopwords = set(stopwords) if stopwords else set()

    def tokenize_stop(self, text):
        words = [word.lower() for word in text.split() if word.isalnum()]
        filtered_words = [word for word in words if word not in self.stopwords]
        
        return filtered_words

    def tokenize(self, text):
        # Пустой список для хранения токенов
        tokens = []
        
        # Инициализируем текущий токен пустой строкой
        token = ""
        delimiters = " \t\n;.,!?()-"
        
        for char in text:
            if char not in delimiters:
                token += char #добавляем символ к текущему токену
            else:
                if token:
                    tokens.append(token) # добавляем токен в список
                    token = ""
                    
        if token:
            tokens.append(token)
            
        return tokens
    
    

stop_words = ["a", "the", "and"]
my_tokenizer = MyTokenizer(stopwords=stop_words)

start_time = timeit.default_timer()

print("Количество токенов со стоп словами", len(my_tokenizer.tokenize_stop(text)))
print("Токены функции стоп слова", my_tokenizer.tokenize_stop(text))

print("Количество токенов ", len(my_tokenizer.tokenize(text)))
print("Токены",my_tokenizer.tokenize(text))




print(timeit.default_timer() - start_time)
```
